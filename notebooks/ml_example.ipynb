```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10)
m_real =1.0
y = m_real*x + np.random.normal(0,0.5,len(x))
guess_error = []
```

```python
guess = 0.1

plt.figure()
plt.scatter(x,y,label='real')
y_guess = guess*x
plt.plot(x,y_guess,label='guess',c='r')
guess_error.append(np.sum(np.abs((y-y_guess))))
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.figure()
plt.bar(np.arange(len(guess_error)),guess_error)
plt.xlabel('iteration')
plt.ylabel('error')
```
